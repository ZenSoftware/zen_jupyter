# **üìò Formal Expression of the Complex Exponential**

If  
$$
z = x + iy,
$$  
where $x = \Re(z)$ and $y = \Im(z)$, then the exponential function is defined by:

$$
\boxed{e^{x+iy} = e^{x}(\cos y + i\sin y)}
$$

This identity follows directly from **Euler‚Äôs formula**:

$$
e^{iy} = \cos y + i\sin y.
$$

So the exponential of a complex number naturally splits into:

- a **real scaling factor** $e^x$, and  
- a **rotation on the unit circle** $\cos y + i\sin y$.

---

# **üß© Why this works**

Start with the power series definition of the exponential:

$$
e^{z} = \sum_{n=0}^{\infty} \frac{z^n}{n!}.
$$

Plug in $z = x + iy$, separate real and imaginary parts, and you recover:

$$
e^{x+iy} = e^x e^{iy} = e^x(\cos y + i\sin y).
$$

This is the standard definition used in complex analysis and differential equations.

---

# **‚ú® Geometric Interpretation**

- $e^x$ stretches the complex plane radially.
- $e^{iy}$ rotates by angle $y$.
- So $e^{x+iy}$ is a **stretch + rotation**.

# üåü **Why All Exponentials Can Be Written Using Euler‚Äôs Number**

Suppose you start with a ‚Äúnormal‚Äù exponential function:

$$
a^x
$$

where $a > 0$ and $a \neq 1$.

The claim is that **every such exponential can be rewritten in the form**:

$$
a^x = e^{kx}
$$

for some constant $k$.  
This is not a coincidence ‚Äî it‚Äôs built into how exponentials behave.

Let‚Äôs unpack why.

---

# üîë 1. **The natural logarithm is the inverse of the exponential**

For positive $a$, the natural logarithm satisfies:

$$
\ln(a) = \text{the unique number such that } e^{\ln(a)} = a.
$$

This identity is the key to everything that follows.

---

# üîë 2. **Rewrite the base using $e$**

Start with:

$$
a = e^{\ln(a)}.
$$

This is true for *every* positive $a$.  
Now raise both sides to the power $x$:

$$
a^x = \left(e^{\ln(a)}\right)^x.
$$

Use the power rule for exponentials:

$$
\left(e^{\ln(a)}\right)^x = e^{x\ln(a)}.
$$

So we get the universal identity:

$$
\boxed{a^x = e^{x\ln(a)}}.
$$

This is the fundamental reason all exponentials can be written using base $e$.

---

# üîë 3. **Interpretation: $\ln(a)$ is the ‚Äúgrowth constant‚Äù**

The number:

$$
k = \ln(a)
$$

is the **continuous growth rate** associated with the base $a$.

So:

$$
a^x = e^{kx}
$$

means:

- $a^x$ grows by a factor of $a$ per unit step  
- $e^{kx}$ grows continuously at rate $k$

These are two different ways of describing the same growth process.

---

# üîë 4. **Why this matters**

### **1. Calculus loves $e$**  
The derivative of $e^{kx}$ is beautifully simple:

$$
\frac{d}{dx} e^{kx} = k e^{kx}.
$$

But the derivative of $a^x$ is:

$$
\frac{d}{dx} a^x = (\ln a)\, a^x.
$$

This formula *comes from* rewriting $a^x$ as $e^{x\ln(a)}$.

### **2. Complex exponentials require $e$**  
Euler‚Äôs formula:

$$
e^{i\theta} = \cos\theta + i\sin\theta
$$

is the gateway to all complex exponential behavior.  
No other base has this property.

### **3. Differential equations prefer $e$**  
Solutions to linear differential equations naturally take the form:

$$
e^{kx}
$$

not $a^x$.  
The $a^x$ form is just a disguise.

---

# üîë 5. **Geometric meaning**

Writing:

$$
a^x = e^{x\ln(a)}
$$

means:

- $a$ tells you the **discrete multiplier**  
- $\ln(a)$ tells you the **continuous rate**

They encode the same growth, just in different languages.

---

# üìù **Summary**

Every exponential function $a^x$ can be rewritten as:

$$
\boxed{a^x = e^{x\ln(a)}}
$$

because:

- $\ln(a)$ is defined as the number satisfying $e^{\ln(a)} = a$
- exponentials convert powers into products
- the base $e$ is the unique number whose exponential has derivative equal to itself
- continuous growth is naturally expressed using $e$

This is why mathematicians overwhelmingly prefer $e$ as the ‚Äútrue‚Äù base of exponentials ‚Äî all others are just $e$ in disguise.
