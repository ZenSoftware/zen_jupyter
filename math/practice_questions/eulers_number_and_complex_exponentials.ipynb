> What is the defining property of the exponential function that uniquely determines $e^x$ among all functions?

$ \frac{d}{dx} e^x = e^x $

More generally, 
- the fact that $ \frac{d}{dx} a^x = ln(a) a^x $
- then if $a = e$
- $ \frac{d}{dx} e^x = ln(e) e^x $
- where $ln(e) = 1$
- thus $ \frac{d}{dx} e^x = e^x $

It is the only base $a$ that makes the coeficient $ln(a)$ of the derivative to equal 1.

> The exponential function with base $e$ is uniquely characterized by the differential equation  
> $$
> f'(x) = f(x), \quad f(0) = 1.

> $$
