# 🧠 **Does breaking charm reset the damage counter?**  
**Yes.**  
On Project Quarm, when charm breaks, the mob’s “damage accounting” resets as if it has just entered combat for the first time.

This is explicitly how the Quarm.Guide Enchanter FAQ describes it, and it’s the foundation of the mez‑blur / charm‑break XP trick.

So after charm breaks:

- The mob’s **current HP becomes its new “full health” baseline**  
- The server begins tracking damage percentages **from that point forward**  
- Any damage your pet did *before* the break **no longer counts**  
- Only damage done **after** the break is used to determine whether you or the pet did >50%

# 🧩 **Why this works**

> The entity (player or pet) that does **more than 50%** of the damage during the current combat window determines the XP split.

Charm break = new combat window.
