# ⚔️ 1. The Classic Rule: **Be on the Hate List = Get XP**  
On classic-era EverQuest (and on P99), the rule was extremely simple:

### ✔️ If you were on the mob’s hate list when it died, you received full XP.  
It didn’t matter if you did:
- 1 damage  
- 0 damage  
- Only debuffed  
- Only healed someone fighting it  
- Only sat there and generated proximity aggro  

As long as you were on the hate list, your pet could do 100% of the damage and you still received full experience.

### 🧩 Why this worked  
Classic EQ treated pets as extensions of the player.  
They did not “steal” XP unless the *pet itself* was the only entity on the hate list.

So the classic rule was:
> **Tag the mob in any way → you get full XP.**

This is why old-school enchanters and mages never worried about damage percentages.

---

# 🔥 2. The Quarm Rule: **Damage Percentage Determines XP**  
Project Quarm intentionally **does not** use the classic hate‑list rule.  
Instead, it uses a modernized, damage-based system:

### ✔️ If your pet does **more than 50%** of the damage from the mob’s out‑of‑combat state to death,  
### ➜ the pet counts as a group member and you receive **50% XP**.

This is explicitly documented in your open tab:  
  [quarm.guide](https://quarm.guide/enchanter-faq)

> “Any pet… that does more than 50% of the damage… will take experience as if it was a member of your group.”

This is a **fundamental departure** from classic mechanics.

# 🎯 3. Why “Do 1 Damage” Is a Misnomer  
The common advice — *“just hit it once so your pet doesn’t steal XP”* — implies that the system works like classic EQ, where tagging the mob is what matters.

But on Quarm:

### ❌ Doing 1 damage does **NOT** prevent pet XP steal.  
### ✔️ Only doing **more than 50% of the damage** prevents it.

You could:
- Nuke for 1 damage  
- Dot for 1 damage  
- Melee for 1 damage  
- Throw a dagger for 1 damage  

…and your pet will still steal XP if it does 51%+ of the total damage.

### The misnomer comes from players confusing:
- **Classic hate‑list tagging**  
with  
- **Quarm’s damage‑percentage threshold**

These are not remotely the same mechanic.

This is the mechanic.  
Everything else — “do 1 damage,” “tag it,” “just hit it once” — is legacy thinking from classic EQ that does not apply.
