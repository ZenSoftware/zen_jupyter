**Improved Invisibility keeps you invisible through actions that would normally break Invisibility.**  
That’s the core difference.

### Functional differences
- **Invisibility** — breaks the moment you cast a spell, click most items, or perform many actions. It’s fragile but lasts longer (12–20 minutes) and can be cast on others.   [pqdi.cc](https://www.pqdi.cc/spell/42)
- **Improved Invisibility** — *does not break when you cast spells or take most actions*. It’s self‑only, shorter duration (fixed 10 minutes), and costs more mana.   [pqdi.cc](https://www.pqdi.cc/spell/1406)

### What both share
- Same invisibility effect type  
- Same “unresistable” flag  
- Same fade behavior (random checks, can drop at any time)

### Why Improved Invis matters in practice
It lets you:
- Buff yourself while staying invisible  
- Begin a pull or cast an opener without popping out  
- Move through dangerous areas without worrying about a random click breaking invis
