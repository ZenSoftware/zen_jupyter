# 🏹 What Disciplines *Are* (in EverQuest Classic/P99)

Disciplines are **short‑duration, long‑cooldown combat abilities** introduced in Kunark and expanded in Velious.  
They’re essentially “super moves” — temporary states that massively boost a specific aspect of your class.

For Rangers, they’re not used constantly like spells; they’re **strategic**, saved for key moments such as:

- Burning down a raid target  
- Surviving a dangerous pull  
- Maximizing bow DPS  
- Avoiding a lethal melee round  

You activate them via:

- **Alt+C** (discipline window), or  
- A macro:  
  ```
  /disc trueshot
  ```

Cooldowns shrink slightly as you level from 51–60.

# 🌲 Ranger Disciplines (Velious Era)

Here’s what *you* get as a Ranger, and when they matter.

## ⭐ Level 51 — **Resistant Discipline**
- **Effect:** +10 to all resistances  
- **Duration:** 1 minute (5 minutes on P99)  
- **Reuse:** 60 minutes  
- **Use it when:** You expect AE spells — dragons, giants, casters, etc.

This is your “oh crap, the dragon is about to breathe” button.

---

## ⭐ Level 54 — **Fearless Discipline**
- **Effect:** Complete immunity to fear  
- **Duration:** 11 seconds  
- **Reuse:** 60 minutes  
- **Use it when:** Fighting mobs with fear (e.g., some undead, dragons, certain raid bosses)

Short but lifesaving. Prevents you from running into adds or off cliffs.

---

## ⭐ Level 55 — **Trueshot Discipline** (Your Signature Ability)
- **Effect:**  
  - +105% bow damage  
  - +12% archery hit chance  
- **Duration:** 2 minutes  
- **Reuse:** 72 minutes  
- **Use it when:**  
  - Raid burn phases  
  - Outdoor fights where you can safely bow  
  - Any situation where you want to feel like Legolas

This is *the* Ranger discipline. It turns your bow into a raid‑level DPS weapon for two minutes.  
With good arrows and a solid bow, Trueshot is your biggest contribution to raid DPS.

---

## ⭐ Level 60 — **Weapon Shield Discipline**
- **Effect:** Parry *all* frontal melee attacks  
- **Duration:** 15 seconds (20 seconds on P99)  
- **Reuse:** 72 minutes  
- **Use it when:**  
  - You pull aggro  
  - You need to tank briefly  
  - You’re saving a wipe  
  - You’re doing a risky pull or CC maneuver

This is your “invulnerability” button.  
It doesn’t reduce spell damage, but it makes you immune to melee from the front.

# 🧭 What You Actually Need to Know as a New Ranger

Here’s the practical summary:

## 🎯 1. You only get **four** disciplines  
Rangers don’t have a huge list — just four, and each has a clear purpose.

## 🎯 2. **Trueshot** is your bread‑and‑butter DPS cooldown  
Use it on raids, named mobs, or any long outdoor fight.

## 🎯 3. **Weapon Shield** is your emergency tank button  
It can save your life or your group’s life.  
Use it confidently — it’s meant for clutch moments.

## 🎯 4. **Resistant** and **Fearless** are situational  
They’re not used every day, but when they matter, they matter *a lot*.

## 🎯 5. Cooldowns are long  
Plan your usage. These aren’t spam abilities.

## 🎯 6. Make macros  
Example:
```
/disc trueshot
/gsay TRUESHOT BURN!
```
