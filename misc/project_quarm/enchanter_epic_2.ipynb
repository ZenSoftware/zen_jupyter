You used a “5 PHs → 1 AM” and “4 AMs → 1 robe” logic, which is exactly the right structure. The only subtle correction is that **expected value doesn’t mean you literally get the 5th PH or the 4th AM** — it’s the average over many trials.

## 📊 What this means in practice  
Expected value is not a guarantee. Realistic outcomes:

- **Lucky:** 1 AM → robe in 40 minutes  
- **Average:** ~160 minutes  
- **Unlucky:** 6–8 AMs → 240–320 minutes  
- **Very unlucky:** 10+ AMs → 400+ minutes  

Classic EverQuest RNG is streaky, so variance is real.
