> Why does [Steve Brunton](https://youtu.be/PjeOmOz9jSY?si=aVunUYfH6NTQ6aHF) use n to denote the number or rows, and m to denote the number of columns.  From what I have seen, it is usually the other way around

The short answer: **Steve Brunton’s choice of $n$ for rows and $m$ for columns is deliberate, consistent within control theory, and tied to how engineers think about measurements vs. variables.**  
It’s not “wrong”—it’s a *discipline‑specific convention* that differs from the one most mathematicians use.

Let me give you the clean, structured breakdown you like.

---

### 🎯 **The core reason**
In **control theory**, **data science**, and **dynamical systems**, the matrix  
$$
A \in \mathbb{R}^{n \times m}
$$
is often interpreted as:

- $n$: **number of measurements**, equations, sensors, or constraints  
- $m$: **number of state variables**, unknowns, or features

This aligns with the canonical linear system:

$$
Ax = b
$$

Where:

- $A$ has **one row per equation** → $n$ equations  
- $A$ has **one column per unknown** → $m$ unknowns

So in Brunton’s world:

- **Rows = equations = measurements = $n$**  
- **Columns = variables = states = $m$**

This is extremely natural for engineers.

---

### 📐 **Why mathematicians often do the opposite**
In pure math and linear algebra textbooks, the convention is usually:

$$
A \in \mathbb{R}^{m \times n}
$$

with:

- $m$: number of rows  
- $n$: number of columns  

This comes from a more abstract tradition where matrices are thought of as linear maps:

$$
A : \mathbb{R}^n \to \mathbb{R}^m
$$

So:

- **Domain dimension = $n$** → number of columns  
- **Codomain dimension = $m$** → number of rows  

This is the “standard” in math departments.

---

### 🧭 **Why Brunton flips it**
Brunton is teaching **applied linear algebra for data**, not abstract linear algebra.

His notation emphasizes:

- **Data points** (rows)  
- **Features** (columns)

And in data science:

- A dataset with $n$ samples and $m$ features is written as  
  $$
  X \in \mathbb{R}^{n \times m}
  $$

This is the **machine learning convention**.

Brunton is consistent with:

- machine learning  
- statistics  
- control theory  
- signal processing  
- system identification  
- regression and least squares literature

So he’s not being idiosyncratic—he’s aligning with the applied world.

---

### 🧩 **The deeper intuition (your style)**
Mathematicians think of matrices as **maps**.  
Engineers think of matrices as **data**.

So the notation flips depending on whether you think:

- “This matrix transforms vectors” → math → $m \times n$  
- “This matrix stores measurements” → engineering → $n \times m$

Both are valid.  
Both are internally consistent.  
They simply encode different mental models.

---

### 🌟 **Summary**
- Brunton uses **$n$ = rows**, **$m$ = columns** because he is thinking in terms of **data and measurements**.  
- Math textbooks use **$m$ = rows**, **$n$ = columns** because they think in terms of **linear maps**.  
- Neither is wrong; they serve different conceptual frameworks.
