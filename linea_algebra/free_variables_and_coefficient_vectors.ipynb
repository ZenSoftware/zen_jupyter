# 🎯 The core idea  
A **free variable** gets a **coefficient vector** because each free variable contributes one independent “direction” in the null space. That direction is encoded as a vector — the coefficient vector.

This is how we build a basis for the null space.

Let’s break it down in your style: geometric, intuitive, and precise.

# 🧱 Step-by-step: Why free variables produce coefficient vectors

When you solve  
$$
A x = 0,
$$  
you end up expressing some variables in terms of others.

- **Leading variables** (pivot variables) are determined by the equations.
- **Free variables** can be chosen freely.

Suppose your solution looks like:

$$
x = 
\begin{pmatrix}
x_1 \\ x_2 \\ x_3 \\ x_4
\end{pmatrix}
=
\begin{pmatrix}
-2s + 3t \\
\;\;s \\
\;\;t \\
0
\end{pmatrix}
$$

Here:

- $s$ and $t$ are **free variables**.
- Each free variable contributes one “direction” in the solution space.

Now rewrite the solution by **splitting it by free variable**:

$$
x = s
\begin{pmatrix}
-2 \\ 1 \\ 0 \\ 0
\end{pmatrix}
+
t
\begin{pmatrix}
3 \\ 0 \\ 1 \\ 0
\end{pmatrix}
$$

Those two vectors are the **coefficient vectors**.

---

# 🌟 What is a coefficient vector?

A **coefficient vector** is the vector that appears when you isolate the contribution of one free variable.

Formally:

> **For each free variable, set that variable = 1 and all other free variables = 0.  
> The resulting solution vector is the coefficient vector.**

It’s the “pure direction” associated with that free variable.

---

# 🧭 Why this matters

The null space is the set of all solutions to $Ax = 0$.  
Every solution is a linear combination of these coefficient vectors.

So:

- Number of free variables = number of coefficient vectors  
- Number of coefficient vectors = dimension of the null space  
- These coefficient vectors form a **basis** for the null space

This is exactly what the video states at [00:07:43](#timestamp-00:07:43)–[00:07:47)   [youtube.com](https://www.youtube.com/watch?v=eeGvVyesafw):

> “Each free variable gets a coefficient vector, and that number of vectors is the dimension of the null space.”

# 🎯 The core idea  
When you solve $Ax = 0$, each free variable gives you one **independent direction** in the null space.  
The **coefficient vector** for that free variable *is* that direction.

All possible null‑space vectors are linear combinations of those coefficient vectors.

So yes:

> **The coefficient vectors span the null space.  
> And because they are independent, they form a basis.**

---

# 🧱 Why this is true (step-by-step)

Suppose your solution to $Ax = 0$ looks like:

$$
x = 
\begin{pmatrix}
-2s + 3t \\
s \\
t \\
0
\end{pmatrix}
$$

Here:

- $s$ and $t$ are free variables.
- The null space is all vectors you get by choosing any $s, t \in \mathbb{R}$.

Now split the solution by free variable:

$$
x = s
\begin{pmatrix}
-2 \\ 1 \\ 0 \\ 0
\end{pmatrix}
+
t
\begin{pmatrix}
3 \\ 0 \\ 1 \\ 0
\end{pmatrix}
$$

Those two vectors are the **coefficient vectors**.

Let’s name them:

$$
v_1 = \begin{pmatrix} -2 \\ 1 \\ 0 \\ 0 \end{pmatrix}, \quad
v_2 = \begin{pmatrix} 3 \\ 0 \\ 1 \\ 0 \end{pmatrix}
$$

Then the null space is:

$$
\text{Null}(A) = \text{span}\{v_1, v_2\}.
$$

That’s the whole story.

---

# 🌟 Why they form a basis  
A basis must satisfy two things:

### 1. **Span**  
Every vector in the null space must be expressible as a combination of the basis vectors.

We already saw:

$$
x = s v_1 + t v_2
$$

So they span the null space.

### 2. **Linear independence**  
No coefficient vector can be written as a combination of the others.

This is guaranteed because:

- each coefficient vector corresponds to a *different* free variable,
- and free variables are independent parameters.

Therefore, the coefficient vectors are automatically linearly independent.

So they form a basis.

---

# 🧭 The deeper intuition (your style)

Think of the null space as a hidden tunnel system inside the matrix — the set of directions that get annihilated.

Each free variable is like a lever you can pull.

- Pull lever $s$: you move along direction $v_1$.  
- Pull lever $t$: you move along direction $v_2$.

The coefficient vectors are the **pure directions** you get when you pull one lever at a time.

All possible movements in the tunnel system are combinations of these pure directions.

That’s exactly what “span” means.

And because each lever controls a unique direction, the coefficient vectors don’t overlap — they’re independent.

That’s exactly what “basis” means.

# 🔚 Summary  
- Yes, coefficient vectors describe the **span** of the null space.  
- They form a **basis** because they are independent and span the space.  
- Number of free variables = number of coefficient vectors = dimension of the null space.
