# üéº Why the $2\pi$ Shows Up (or Disappears)

The Fourier transform fundamentally converts between:

- **Time domain** $t$  
- **Frequency domain** $f$ or **angular frequency** $\omega$

The key distinction is:

### üîπ If you measure frequency in **cycles per second** (Hz), the natural oscillation is  
$$
e^{-2\pi i f t}
$$

### üîπ If you measure frequency in **radians per second**, the natural oscillation is  
$$
e^{-i \omega t}
$$

And since  
$$
\omega = 2\pi f,
$$  
the two formulas are literally the same transform written in different units.

---

# üìê Three Common Conventions

## **1. Engineering / Signal Processing Convention (uses Hz)**
$$
\mathcal{F}\{x(t)\}(f) = \int_{-\infty}^{\infty} x(t)\, e^{-2\pi i f t}\, dt
$$

- Uses **cycles per second**  
- The $2\pi$ is in the exponent  
- Inverse transform has no $2\pi$

This matches how oscilloscopes, audio signals, and DSP systems think about frequency.

---

## **2. Physics Convention (uses angular frequency)**
$$
\mathcal{F}\{x(t)\}(\omega) = \int_{-\infty}^{\infty} x(t)\, e^{-i \omega t}\, dt
$$

- Uses **radians per second**  
- No $2\pi$ in the exponent  
- Inverse transform has a factor $1/(2\pi)$

Physicists prefer angular frequency because it plays nicely with differential equations.

---

## **3. Pure Mathematics Convention (symmetrized)**
$$
\mathcal{F}\{x(t)\}(\xi) = \frac{1}{\sqrt{2\pi}} \int x(t)\, e^{-i \xi t}\, dt
$$

- Splits the $2\pi$ evenly between forward and inverse transforms  
- Makes the transform **unitary** on $L^2$  
- Beautiful for functional analysis and PDEs

# üéØ Why no single ‚Äúcorrect‚Äù definition exists

Because the Fourier transform is not a physical law ‚Äî it‚Äôs a *tool*.  
Different fields optimize for different properties:

| Field | Preferred Form | Reason |
|------|----------------|--------|
| Engineering | $e^{-2\pi i f t}$ | Frequency measured in Hz |
| Physics | $e^{-i \omega t}$ | Angular frequency simplifies equations |
| Mathematics | Symmetric $1/\sqrt{2\pi}$ | Makes the transform unitary |

They‚Äôre all the same transform up to a change of variables.

# üß† The intuition that ties it all together

The Fourier transform is fundamentally about decomposing a function into oscillations.  
The only choice is:

- Do you count oscillations per **cycle** (Hz)?  
- Or per **radian** (angular frequency)?

That single choice determines where the $2\pi$ lands.
