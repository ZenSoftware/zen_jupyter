# üåü The Major Types of Integrals (with Names)

## 1. **Indefinite Integral**
- **Form:** $\displaystyle \int f(x)\,dx$
- **Meaning:** The *antiderivative* of $f(x)$; a family of functions.
- **Output:** A function $F(x) + C$
- **Use:** Reversing differentiation.

---

## 2. **Definite Integral**
- **Form:** $\displaystyle \int_a^b f(x)\,dx$
- **Meaning:** Net signed area under the curve from $a$ to $b$
- **Output:** A number
- **Use:** Area, accumulation, total change.

---

# üîç How Integrals Are Classified

## A. **By Domain / Limits**

### 3. **Improper Integral**
- **Form:** Infinite limits or unbounded integrand  
  $\displaystyle \int_0^\infty e^{-x}\,dx$
- **Meaning:** Limit of definite integrals
- **Use:** Probability, physics, Fourier transforms.

---

### 4. **Multiple Integrals**
- **Double integral:** $\displaystyle \iint_R f(x,y)\,dA$  
- **Triple integral:** $\displaystyle \iiint_V f(x,y,z)\,dV$
- **Use:** Volume, mass, probability densities.

---

### 5. **Line Integral**
- **Scalar line integral:** $\displaystyle \int_C f(x,y)\,ds$  
- **Vector line integral:** $\displaystyle \int_C \vec{F}\cdot d\vec{r}$
- **Use:** Work done by a force field, circulation.

---

### 6. **Surface Integral**
- **Scalar:** $\displaystyle \iint_S f\,dS$  
- **Vector:** $\displaystyle \iint_S \vec{F}\cdot d\vec{S}$
- **Use:** Flux through a surface.

---

### 7. **Volume Integral**
- Essentially a triple integral over a 3D region.

---

# B. **By Coordinate System**

### 8. **Polar Integral**
$$
\int\!\!\int_R f(r,\theta)\, r\,dr\,d\theta
$$

### 9. **Cylindrical Integral**
$$
\iiint f(r,\theta,z)\, r\,dr\,d\theta\,dz
$$

### 10. **Spherical Integral**
$$
\iiint f(\rho,\phi,\theta)\,\rho^2\sin\phi\, d\rho\, d\phi\, d\theta
$$

---

# C. **By Purpose / Interpretation**

### 11. **Riemann Integral**
- Based on Riemann sums.
- Most common in calculus.

### 12. **Riemann‚ÄìStieltjes Integral**
$$
\int_a^b f(x)\,dg(x)
$$
- Generalizes Riemann by integrating with respect to another function.

### 13. **Lebesgue Integral**
- Integrates with respect to measure.
- Foundation of modern probability and analysis.

### 14. **Path Integral (Quantum Mechanics)**
- Feynman integrals over infinite-dimensional spaces.

---

# D. **Specialized Integrals**

### 15. **Contour Integral (Complex Analysis)**
$$
\oint_\gamma f(z)\,dz
$$
- Integral over a curve in the complex plane.

### 16. **Fourier Integral**
- Used to decompose functions into frequencies.

### 17. **Laplace Integral**
- Basis of the Laplace transform.

### 18. **Cauchy Principal Value**
- A way to assign values to divergent integrals.

---
