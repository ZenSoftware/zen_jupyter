A **sinusoid** is any signal of the form:

It is a **single**, pure tone with:

**Harmonics** are a *set* of sinusoids whose frequencies are integer multiples of a fundamental frequency.

If the fundamental frequency is $f_0$, then the harmonics are:

$$
f_1 = f_0,\quad f_2 = 2f_0,\quad f_3 = 3f_0,\quad \dots
$$

Harmonics only make sense **relative to a fundamental**.
