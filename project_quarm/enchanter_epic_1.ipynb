## 🧭 High‑Level Differences Between P99 and Project Quarm

### 1. **Jeb’s Seal Is *Required* on Project Quarm**
On P99, older versions of the quest allowed you to skip Jeb’s Seal or get sacks from any master without it.  
On PQ, the P99 page explicitly notes a patch change:

- **“Jeb’s Seal (This is no longer optional as of 2025‑12‑21 patch)”**   [wiki.project1999.com](https://wiki.project1999.com/Enchanter_Epic_Quest)

This means:
- You *must* do Stofo → Notes → Jeb → Seal before any master will cooperate.
- You cannot pre‑farm sacks or skip steps.

This is the single biggest mechanical difference.

---

## 👜 2. Bag Mechanics Are Slightly Different
P99 allows:
- Getting a sack from *any* master  
- Using that sack to perform *any* combine  
- (This was a quirk of the original quest.)

PQDI’s Jeb page shows:
- Jeb gives **Jeb’s Seal** and instructs you to show it to each master  
- Masters give you a sack tied to their step  
- PQDI does *not* mention cross‑compatibility of sacks, and PQ tends to follow intended-era logic

This strongly suggests:
- **On PQ, sacks are intended to be used for their specific master’s combine.**  
- Cross‑use may not work or may be intentionally fixed.

# 🧭 Optimal Project Quarm Enchanter Epic Route  
*(Fastest, safest, least backtracking, and aligned with PQ mechanics)*

This route assumes you want to minimize downtime, avoid contested bottlenecks, and use Luclin-era travel to your advantage.

---

## 1. **Start in Erudin → Stofo → Jeb**  
This locks in the one PQ‑specific requirement: **Jeb’s Seal**.

1. Talk to **Stofo Olan** in Erudin.  
2. Deliver the notes to **a sarnak imitator** in Burning Woods.  
3. Receive **Jeb’s Seal** (confirmed in your PQDI tab: “Take this seal… you will need to show my seal to the other masters…”   [pqdi.cc](https://www.pqdi.cc/script-entities/burningwood/a_sarnak_imitator)).

**Why this matters on PQ:**  
On P99, you could skip this step early. On PQ, you *cannot* progress without the seal.

---

## 2. **Master 1: Modani Qu’Loni (Illusions Master)**  
Jeb’s dialogue tells you to seek Modani first:  
“Modani is a master of illusions… seek him out first.”   [pqdi.cc](https://www.pqdi.cc/script-entities/burningwood/a_sarnak_imitator)

**Location:** Oasis → wandering dark elf illusion.  
**Task:** Kill the phantasm, loot the item, hand in with Jeb’s Seal, get **Enchanted Staff Piece #1**.

**PQ advantage:**  
Spawn behavior is more stable than P99’s erratic pathing. Expect less waiting.

---

## 3. **Master 2: Mizzle Gepple (Enlightened Magic Master)**  
Jeb’s dialogue:  
“Mizzle is a gnome… go to him once you have spoken to Modani.”   [pqdi.cc](https://www.pqdi.cc/script-entities/burningwood/a_sarnak_imitator)

**Location:** Ak’Anon.  
**Task:** Classic “gather components → combine in sack” step.

**PQ advantage:**  
Sacks are likely master‑specific (PQDI implies intended-era logic), so do this step immediately after Modani to avoid mixing sacks.

---

## 4. **Master 3: Nadia Starfeast (Charm Master)**  
Jeb’s dialogue:  
“Nadia… traveled with her relatives to the new land… the outpost is well defended now.”   [pqdi.cc](https://www.pqdi.cc/script-entities/burningwood/a_sarnak_imitator)

**Location:** Firiona Vie.  
**Task:** Kill the phantasm, hand in with Jeb’s Seal, get **Enchanted Staff Piece #3**.

**PQ advantage:**  
FV is much safer on PQ than on P99. Guards won’t randomly obliterate you mid‑turn‑in.

## 5. **Master 4: Polzin Mrid (Phantasms Master)**  
Jeb’s dialogue:  
“Polzin is native to Erudin… after the incident near Paineel, no one has heard from him.”   [pqdi.cc](https://www.pqdi.cc/script-entities/burningwood/a_sarnak_imitator)

**Location:** The Hole / Paineel area.  
**Task:** Kill the phantasm, hand in with Jeb’s Seal, get **Enchanted Staff Piece #4**.

**PQ advantage:**  
The Hole is less contested than P99. You can often walk in and get your kill without competition.

## 6. **Combine the Four Staff Pieces**  
Each master gives you a sack. On PQ, sacks are intended to be used for their specific combine (PQDI text emphasizes showing the seal to each master to receive “the means to bundle them”   [pqdi.cc](https://www.pqdi.cc/script-entities/burningwood/a_sarnak_imitator)).

**Do this:**  
- Combine each staff piece in its correct sack.  
- Combine the resulting four items into **A Bundle of Staves**.

---

## 7. **Return to Jeb in Burning Woods**  
Turn in **A Bundle of Staves** *and* ensure your faction is **Indifferent or better**.

PQDI explicitly states:  
“Faction is > Indifferent” is required for the final turn-in.   [pqdi.cc](https://www.pqdi.cc/script-entities/burningwood/a_sarnak_imitator)

Illusion + Collaboration will get you there.

You receive: **Staff of the Serpent**.
