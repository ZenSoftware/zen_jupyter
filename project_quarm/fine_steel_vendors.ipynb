You can buy **Fine Steel weapons from many standard city merchants**, but the key detail is that **they are not consistently stocked**—they appear only when other players sell them to vendors. That means you won’t find a guaranteed “Fine Steel vendor,” but you *can* reliably shop in places where players frequently dump FS drops.

To anchor this with sourced info: Fine Steel weapons are **not crafted** and are primarily **mob drops** (guards, giants, mid‑ to high‑level mobs) rather than merchant-stocked items. Vendors only have them when players sell them there.   [Project 1999 Wiki](https://wiki.project1999.com/Fine_Steel_Weapons)

## 🗺️ Where you can reliably buy Fine Steel on Project Quarm
These locations work because players constantly sell FS there, so the vendors almost always have a selection of 1H weapons for your charmed pet.

### 1. **North Freeport – Militia House & surrounding vendors**
Players kill Freeport guards constantly, and FS weapons get dumped on the nearest merchants.  
You’ll often find:
- Fine Steel Long Swords  
- Fine Steel Rapiers  
- Fine Steel Warhammers  
- Fine Steel Daggers  

### 2. **East Commonlands Tunnel merchants**
The EC tunnel is the universal dumping ground. Any merchant near the tunnel entrance or inside the tunnel often has:
- FS Scimitars  
- FS Spears  
- FS Short Swords  

### 3. **High Keep – Guard vendors**
High Keep guards drop tons of Fine Steel, and players farming them sell directly to:
- **Merchant Dyllin**  
- **Merchant Breya**  
- **Merchant Katha Firespinner**  
These vendors frequently have a full spread of FS 1H weapons.

### 4. **Qeynos – North Qeynos gate merchants**
Qeynos guards drop FS, and players farming faction or cash often sell to the nearest merchants.

### 5. **Kaladim / Butcherblock dwarf vendors**
Dwarves kill goblins, players kill guards, and FS gets sold to:
- **Gunlok Jure** area merchants  
- **Butcherblock docks** merchants

## 🪓 Best Fine Steel weapons for charm pets
Since you’re using these for your Enchanter charm pet on Quarm, prioritize:

- **Fine Steel Long Sword** (fast, solid ratio)  
- **Fine Steel Rapier** (fastest 1H FS)  
- **Fine Steel Scimitar** (good balance)  
- **Fine Steel Warhammer** (slower, but good damage)  

All of these are common vendor trash and cheap.

---

## 💡 Practical buying strategy
Because FS is vendor trash, the trick is to **check vendors near popular farming spots**:

- Guards → city vendors  
- Giants → EC/WC vendors  
- High Keep guards → HK vendors  
- Mistmoore → LFay vendors  
- Oasis → dock vendors  

If you want the most consistent supply, **High Keep** and **North Freeport** are the best bets.
