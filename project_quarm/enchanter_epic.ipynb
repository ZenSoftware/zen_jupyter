## 🧭 High‑Level Differences Between P99 and Project Quarm

### 1. **Jeb’s Seal Is *Required* on Project Quarm**
On P99, older versions of the quest allowed you to skip Jeb’s Seal or get sacks from any master without it.  
On PQ, the P99 page explicitly notes a patch change:

- **“Jeb’s Seal (This is no longer optional as of 2025‑12‑21 patch)”**   [wiki.project1999.com](https://wiki.project1999.com/Enchanter_Epic_Quest)

This means:
- You *must* do Stofo → Notes → Jeb → Seal before any master will cooperate.
- You cannot pre‑farm sacks or skip steps.

This is the single biggest mechanical difference.

---

## 👜 2. Bag Mechanics Are Slightly Different
P99 allows:
- Getting a sack from *any* master  
- Using that sack to perform *any* combine  
- (This was a quirk of the original quest.)

PQDI’s Jeb page shows:
- Jeb gives **Jeb’s Seal** and instructs you to show it to each master  
- Masters give you a sack tied to their step  
- PQDI does *not* mention cross‑compatibility of sacks, and PQ tends to follow intended-era logic

This strongly suggests:
- **On PQ, sacks are intended to be used for their specific master’s combine.**  
- Cross‑use may not work or may be intentionally fixed.

---

## 🧍‍♂️ 3. NPC Behavior and Dialogue Are Slightly Modernized on PQ
PQDI’s Jeb page includes:
- Cleaned-up dialogue  
- Explicit faction checks  
- Explicit item return logic  
- Modern database notes (2026-01-01)   [pqdi.cc](https://www.pqdi.cc/script-entities/burningwood/a_sarnak_imitator)

P99’s page includes:
- Legacy-era quirks  
- Some broken or inconsistent dialogue  
- Notes about bugs or oddities that may not exist on PQ

This means:
- **PQ is more stable and less buggy** in terms of turn-ins and dialogue triggers.
- You should expect fewer “P99isms” like broken text, missing responses, or inconsistent sack behavior.

---

## 🧟 4. Spawn Behavior Differences (Likely but Not Fully Documented)
PQDI’s page for Jeb does *not* list spawn windows for Verina, Vessel, Felia, etc., but PQ generally uses:
- **Classic-era spawn windows**, but  
- **Reduced variance** compared to P99’s extreme 2–7 day windows

P99’s page explicitly calls out:
- Verina Tomb: 2–7 day window  
- Vessel Drozlin: same window  
- Tangrin: rare  
- Felia Goldenwing: rare  
- Spectral Librarian: PH cycles

PQ tends to:
- Use more reasonable respawn windows  
- Avoid multi-day bottlenecks  
- Reduce variance to keep progression flowing

So while not explicitly stated in your PQDI tabs, **expect PQ to be less punishing than P99 for bottleneck mobs**.

---

## 🧪 5. Faction Requirements Are Slightly Different
PQDI’s Jeb page shows:
- **Faction > Indifferent** required for final turn-in (Bundle of Staves)   [pqdi.cc](https://www.pqdi.cc/script-entities/burningwood/a_sarnak_imitator)

P99’s page:
- Does not explicitly list faction requirements for Jeb  
- Historically, Jeb accepted turn-ins at indifferent or better, but P99 sometimes has hidden faction quirks

PQ’s explicit faction check means:
- Make sure you’re at least indifferent before the final hand-in  
- Illusions + Collaboration should be enough

---

## 🧩 6. PQDI Uses Modern Item Return Logic
PQDI page states:
- “This NPC should return incorrect items given.”   [pqdi.cc](https://www.pqdi.cc/script-entities/burningwood/a_sarnak_imitator)

P99:
- Many epic NPCs *eat* items if you mess up  
- Some have inconsistent return logic

On PQ:
- **You’re safer from catastrophic mis-turn-ins**  
- Still don’t test your luck with multi-questing

---

## 🗺️ 7. Zone Layout Differences (Luclin-era PQ vs. Kunark-only P99)
This isn’t in the two tabs, but it matters for your run:

- PQ has Luclin zones active  
- P99 epic is designed for a Kunark/Velious world  
- Travel, bind points, and faction routes are easier on PQ

This doesn’t change the quest steps, but it changes:
- How fast you can move between steps  
- How safe certain routes are  
- How easy it is to recover from wipes or charm breaks
