## 🧪 3. **What Neuroscience Methods You’d Use**

Here’s where your open Wu Tsai tab becomes relevant — modern neuroscience is obsessed with **seeing the brain in action**.  
Your study wouldn’t need transparent mice, of course, but the same principles apply: **measure neural activity while the participant performs a task**.

### **Common human‑compatible methods:**

| Method | What it tells you | Why it’s useful for gamers |
|-------|-------------------|----------------------------|
| **EEG** | Millisecond‑level timing of brain activity | Tracks attention, prediction, error signals |
| **fMRI** | Which brain regions activate | Maps networks used during high‑level play |
| **Eye‑tracking** | Visual strategy | Shows how experts scan vs. novices |
| **Physiology (HRV, GSR)** | Stress + arousal | Measures flow, pressure, tilt |
| **Behavioral telemetry** | Reaction time, accuracy | Links brain activity to performance |

You can combine these into a **multimodal performance profile**.
