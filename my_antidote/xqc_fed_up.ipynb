Here’s the clean, real‑world distinction:

## **Takeaway:**  
A **jail** holds people for short periods; a **penitentiary** holds people for long-term felony sentences.

---

## 🏢 **Jail**
- **Short-term detention**
- Run by **local** governments (city or county)
- Holds:
  - People awaiting trial
  - People serving **less than 1 year**
  - People awaiting transfer to another facility
- Environment is more transient, with frequent movement in and out

Think of it as the *front door* of the criminal justice system.

---

## 🏛️ **Penitentiary**
- A type of **prison**, specifically for **long-term incarceration**
- Run by **state or federal** governments
- Holds:
  - People convicted of **serious crimes (felonies)**
  - Sentences typically **1 year or more**
- More structured, long-term environment with programs, work assignments, etc.

The word comes from **“penitence”** — originally meant to encourage reflection and reform.

---

## Quick comparison

| Feature | Jail | Penitentiary |
|--------|------|---------------|
| Length of stay | Short-term | Long-term |
| Who runs it | Local (city/county) | State or federal |
| Who stays there | Pre-trial, misdemeanors | Felons, long sentences |
| Purpose | Holding | Punishment + rehabilitation |
