# Why is the exponential distribution memoryless
### 1. Set up the exponential distribution

Take an exponential distribution with rate parameter $\lambda > 0$:

- **PDF**:

$$
f(t) = \lambda e^{-\lambda t}, \quad t \ge 0
$$

- **CDF**:

$$
F(t) = P(T \le t) = 1 - e^{-\lambda t}, \quad t \ge 0
$$

- **Survival function** (tail probability):

$$
S(t) = P(T > t) = 1 - F(t) = e^{-\lambda t}, \quad t \ge 0
$$



---

### 2. Definition of the hazard function

For a continuous nonnegative random variable $T$, the **hazard function** is

$$
h(t) = \frac{f(t)}{S(t)}
$$

where:

- $f(t)$ is the density at time $t$,
- $S(t) = P(T > t)$ is the probability of surviving beyond $t$.

This is the **instantaneous failure rate at time $t$, given survival up to $t$**.

---

### 3. Plug in exponential $f(t)$ and $S(t)$

Now substitute the exponential formulas:

$$
f(t) = \lambda e^{-\lambda t}, \quad S(t) = e^{-\lambda t}
$$

Then

$$
h(t) = \frac{f(t)}{S(t)} 
     = \frac{\lambda e^{-\lambda t}}{e^{-\lambda t}} 
     = \lambda
$$

So the hazard function is

$$
h(t) = \lambda \quad \text{for all } t \ge 0
$$
