# Why is the exponential distribution memoryless
### 1. Set up the exponential distribution

Take an exponential distribution with rate parameter $\lambda > 0$:

- **PDF**:

$$

$$

- **CDF**:

$$

$$

- **Survival function** (tail probability):

$$
S(t) = P(T > t) = 1 - F(t) = e^{-\lambda t}, \quad t \ge 0
$$

### 2. Definition of the hazard function

For a continuous nonnegative random variable $T$, the **hazard function** is

$$

$$

- $f(t)$ is the density at time $t$,
