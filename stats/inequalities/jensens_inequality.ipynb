## üéØ **Core Idea**
Jensen‚Äôs Inequality is a fundamental result in convex analysis and probability.  
It tells you how a **convex function** interacts with **averages**.

At its heart:

> **A convex function of an average is ‚â§ the average of the convex function.**

This single sentence explains why expectations ‚Äúpush outward‚Äù through convex functions.

---

# üìê **Formal Statement**

Let $ f $ be a **convex** function on an interval $ I \subseteq \mathbb{R} $.  
Let $ X $ be a random variable taking values in $ I $.

Then:

$$
f\big( \mathbb{E}[X] \big) \le \mathbb{E}[f(X)].
$$

If $ f $ is **strictly convex**, equality holds **only** when $ X $ is almost surely constant.

---

# üß† **Intuition**

Convex functions ‚Äúcurve upward.‚Äù  
Because of this curvature:

- When you take an average **first**, you stay near the bottom of the curve.
- When you apply the convex function **first**, you move up the curve, then average‚Äîending up higher.

A picture helps: convex functions lie **below** their chords.  
Jensen‚Äôs inequality is the expectation version of that geometric fact.

---

# üßÆ **Discrete Version**

For weights $ \lambda_i \ge 0 $ with $ \sum_i \lambda_i = 1 $:

$$
f\left( \sum_{i=1}^n \lambda_i x_i \right)
\le
\sum_{i=1}^n \lambda_i f(x_i).
$$

This is the deterministic version‚Äîno probability needed.

---

# üìä **Concave Version**

If $ f $ is **concave**, the inequality reverses:

$$
f\big( \mathbb{E}[X] \big) \ge \mathbb{E}[f(X)].
$$

Examples of concave functions:  
- $ \log x $  
- $ \sqrt{x} $

---

# üîç **Examples**

### 1. **Arithmetic‚ÄìGeometric Mean Inequality (AM‚ÄìGM)**  
Using $ f(x) = -\log x $, which is convex:

$$
-\log\left( \mathbb{E}[X] \right)
\le
\mathbb{E}[-\log X]
\quad\Rightarrow\quad
\mathbb{E}[X] \ge \exp(\mathbb{E}[\log X]).
$$

This is exactly AM ‚â• GM.

---

### 2. **Variance is Non‚ÄëNegative**

Take $ f(x) = x^2 $, convex:

$$
(\mathbb{E}[X])^2 \le \mathbb{E}[X^2]
\quad\Rightarrow\quad
\mathbb{E}[X^2] - (\mathbb{E}[X])^2 \ge 0.
$$

---

### 3. **Risk Aversion in Economics**

Utility functions are concave.  
Jensen says:

$$
U(\mathbb{E}[X]) \ge \mathbb{E}[U(X)],
$$

meaning a risk‚Äëaverse agent prefers the utility of the expected payoff over the expected utility of a risky payoff.

---

# üß© **When Does Equality Hold?**

For convex $ f $:

$$
f(\mathbb{E}[X]) = \mathbb{E}[f(X)]
$$

**iff**  
- $ X $ is constant almost surely, or  
- $ f $ is linear on the support of $ X $.

---

# üõ†Ô∏è **How to Prove It (Sketch)**

### Geometric Proof
- A convex function lies below its tangent lines.
- For any $ x $:

$$
f(x) \ge f(a) + f'(a)(x - a).
$$

- Take expectations and choose $ a = \mathbb{E}[X] $.

### Probabilistic Proof
- Approximate $ X $ by simple random variables.
- Apply the discrete version.
- Use limits.
