## üéØ **What It Says**
Markov‚Äôs inequality gives an **upper bound** on the probability that a **non‚Äënegative** random variable takes on a value at least as large as some threshold.

For a non‚Äënegative random variable $X$ and any $a > 0$:

$$
\Pr(X \ge a) \le \frac{\mathbb{E}[X]}{a}
$$

It‚Äôs a very general inequality‚Äîno assumptions about the distribution except non‚Äënegativity.

---

# üß† **Intuition**
Think of $\mathbb{E}[X]$ as the ‚Äúbudget‚Äù of the random variable.

If the average value is small, then the probability of seeing a very large value must also be small‚Äîotherwise the average would be pulled upward.

Markov‚Äôs inequality formalizes this idea.

---

# üìå **Key Conditions**
- $X \ge 0$ almost surely  
- $a > 0$  
- Only the expectation $\mathbb{E}[X]$ is required‚Äîno variance, no distribution shape

---

# üìê **Why It‚Äôs Useful**
- Works even when you know almost nothing about the distribution  
- Forms the foundation for more powerful inequalities:
  - Chebyshev‚Äôs inequality
  - Chernoff bounds
  - Concentration inequalities in probability theory and machine learning

---

# üß© **Proof Sketch (Simple and Elegant)**
Start with the indicator function:

$$
X \ge a \cdot \mathbf{1}_{\{X \ge a\}}
$$

Take expectations:

$$
\mathbb{E}[X] \ge a \cdot \mathbb{E}[\mathbf{1}_{\{X \ge a\}}]
$$

But:

$$
\mathbb{E}[\mathbf{1}_{\{X \ge a\}}] = \Pr(X \ge a)
$$

So:

$$
\Pr(X \ge a) \le \frac{\mathbb{E}[X]}{a}
$$

---

# üìä **Example**
Let $X$ be the time (in minutes) you wait for a bus. Suppose:

$$
\mathbb{E}[X] = 10
$$

What‚Äôs the maximum possible probability that you wait at least 30 minutes?

$$
\Pr(X \ge 30) \le \frac{10}{30} = \frac{1}{3}
$$

Even without knowing the distribution, you know the tail probability can‚Äôt exceed 1/3.

---

# üîç **Tightness**
Markov‚Äôs inequality can be **loose**, but it is **tight** in the sense that there exist distributions where equality holds.

Example:  
Let $X = a$ with probability $\frac{\mathbb{E}[X]}{a}$ and $0$ otherwise.

---

# üß± **Generalizations**
### 1. **For functions of random variables**
If $g(X)$ is non‚Äënegative:

$$
\Pr(g(X) \ge a) \le \frac{\mathbb{E}[g(X)]}{a}
$$

### 2. **Chebyshev‚Äôs inequality**  
Apply Markov to $g(X) = (X - \mu)^2$.

---

# üß≠ **When to Use Markov‚Äôs Inequality**
- You only know the expectation  
- You need a quick, rough upper bound  
- You‚Äôre working with heavy‚Äëtailed distributions  
- You‚Äôre building toward more refined concentration results
