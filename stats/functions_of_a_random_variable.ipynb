# ðŸŽ¯ **1. CDF Transformation**

You start with:

$$
Y = aX + b,\quad X \sim N(\mu,\sigma^2)
$$

Then:

$$
F_Y(y) = P(Y < y) = P(aX + b < y)
$$

Now the key step:

- If $a > 0$:

$$
P(aX + b < y) = P\!\left(X < \frac{y-b}{a}\right)
$$

- If $a < 0$, the inequality flips:

$$
P(aX + b < y) = P\!\left(X > \frac{y-b}{a}\right)
$$

So the fully correct expression is:

$$
F_Y(y) =
\begin{cases}
F_X\!\left(\frac{y-b}{a}\right), & a>0 \\
1 - F_X\!\left(\frac{y-b}{a}\right), & a<0
\end{cases}
$$

Most textbooks assume $a>0$ for simplicity, but itâ€™s good to know the general case.

$$

Derivative:

$$
f_Y(y) = \frac{1}{|a|} f_X\!\left(\frac{y-b}{a}\right)
$$

The absolute value is important â€” it ensures the PDF stays nonâ€‘negative even when $a<0$.

# ðŸŽ¯ **3. Distribution of $Y$**

Now the punchline.

A linear transformation of a normal random variable is still normal.  
Compute the mean and variance:

### Mean:
$$
E[Y] = E[aX + b] = aE[X] + b = a\mu + b
$$

### Variance:
$$
\mathrm{Var}(Y) = \mathrm{Var}(aX + b) = a^2 \mathrm{Var}(X) = a^2 \sigma^2
$$

So yes:

$$
\boxed{Y \sim N(a\mu + b,\; a^2\sigma^2)}
$$

This holds for **any real $a$** (positive or negative).

# ðŸŒŸ **4. Why this works intuitively**

A normal distribution is completely determined by:

- where itâ€™s centered (mean)
- how spread out it is (variance)

Multiplying by $a$:

- stretches or compresses the distribution by $|a|$
- flips it if $a<0$

Adding $b$:

- shifts the whole curve left or right

These operations preserve the bell shape â€” so the result is still normal.

# ðŸŽ¯ **Why does $\mathrm{Var}(aX + b) = a^2 \mathrm{Var}(X)$?**

Start from the definition:

$$
\mathrm{Var}(Y) = \mathbb{E}\!\left[(Y - \mathbb{E}[Y])^2\right]
$$

Let $Y = aX + b$. Then:

$$
\mathrm{Var}(aX + b)
= \mathbb{E}\!\left[(aX + b - \mathbb{E}[aX + b])^2\right]
$$

Now use linearity of expectation:

$$
\mathbb{E}[aX + b] = a\mathbb{E}[X] + b
$$

Substitute:

$$
\mathrm{Var}(aX + b)
= \mathbb{E}\!\left[(aX + b - (a\mathbb{E}[X] + b))^2\right]
$$

The $b$'s cancel:

$$
= \mathbb{E}\!\left[(aX - a\mathbb{E}[X])^2\right]
$$

Factor out $a$:

$$
= \mathbb{E}\!\left[a^2 (X - \mathbb{E}[X])^2\right]
$$

Pull the constant $a^2$ outside the expectation:

$$
= a^2 \mathbb{E}\!\left[(X - \mathbb{E}[X])^2\right]
$$

But the remaining expectation is exactly the definition of $\mathrm{Var}(X)$:

$$
\boxed{\mathrm{Var}(aX + b) = a^2 \mathrm{Var}(X)}
$$

# ðŸŒŸ **Intuition**

- Adding $b$ shifts the distribution left or right.  
  **Shifts do not change spread**, so variance is unaffected.

- Multiplying by $a$ stretches or compresses the distribution.  
  Stretching by a factor of $|a|$ multiplies distances from the mean by $|a|$.  
  Variance measures *squared* distances, so it scales by $a^2$.

This is why the sign of $a$ doesnâ€™t matter â€” flipping the distribution doesnâ€™t change its spread.
