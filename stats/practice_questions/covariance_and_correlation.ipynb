### **1. Definitions & Intuition**
> What does covariance measure, in plain language?

~~Covariance is a measure of variance of joint distributions. It can categorize distributions into one of three categories.~~

Covariance is a measure of how variables vary together.  It tells you whether or not increases in one variable tend to be associated with increases or decreases in the other.

- Positive covariance: the variables tend to move in the same direction
- Negative covariance: the variables tend to move in opposite directions
- ~~Zero covariance meaning that there is no relation between the two variables~~ Zero covariance: no linear relationship in how they move together (though there could possibly be non-linear relationships that exist)

**In plain language:**

Covariance captures whether or not two variables "move together", "move oppositely" or "don't move together in any consitent linear way"
