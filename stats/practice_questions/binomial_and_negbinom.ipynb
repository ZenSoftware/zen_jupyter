### **1.**
A machine produces items, each independently defective with probability $0.02$. You inspect $40$ items.  
What is the probability that exactly $3$ are defective?

```python
import scipy.stats as stats

x = 3
p = 0.02
n = 40
stats.binom.pmf(x, n, p)
```

### **2.**
A gamer lands a critical hit with probability $0.25$ on each attack.  
What is the probability that the 4th critical hit occurs on the 12th attack?

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

r = 4
n = 12
p = 0.25

x_vals = np.arange(r, r + 50)
pmf = stats.nbinom.pmf(x_vals - r, r, p)
answer = stats.nbinom.pmf(n- r, r, p)
plt.bar(x_vals, pmf)
plt.scatter([n], [answer], color="r", label=f"P(X={n}) = {answer:.3}")
plt.axvline(n, linestyle="--", color="r")
plt.title(f"Negative Binomial PMF (r = {r}, p = {p})")
plt.xlabel("Number of trials")
plt.ylabel("Probability")
plt.legend()
plt.tight_layout()
plt.show()
```

### **3.**
A salesperson closes a deal with probability $0.1$ per customer.  
What is the probability they need at most 20 customers to make 3 sales?

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

p = 0.1
r = 3
n = 20

x_vals = np.arange(r, r + 50)
answer = stats.nbinom.cdf(n - r, r, p)
pmf = stats.nbinom.pmf(x_vals - r, r, p)
colors = ['C1' if x <= n else 'C0' for x in x_vals]
plt.bar(x_vals, pmf, color=colors)
plt.axvline(n + 0.5, color="C1", linestyle="--", label=f"P(X <= {n}) = {answer:.3}")
plt.title(f"Negative Binomial PMF (r = {r}, p = {p})")
plt.xlabel("Number of customers")
plt.ylabel("Probability")
plt.legend()
plt.tight_layout()
plt.show()
```

### **4.**
A student guesses on a 15â€‘question multipleâ€‘choice quiz, each with 4 options.  
What is the probability they get at least 5 correct?

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

p = 0.25
n = 15
k = 5
x_vals = np.arange(n+1)

answer = stats.binom.pmf(k, n, p)
pmf = stats.binom.pmf(x_vals, n, p)
colors = ['C1' if x <= k else 'C0' for x in x_vals]
plt.bar(x_vals, pmf, color=colors)
plt.axvline(k+0.5, color="C1", linestyle="--", label=f"P(X <= {k}) = {answer:.3}")
plt.xticks(x_vals)
plt.title(f"Binomial(n = {n}, p = {p})")
plt.xlabel(f"Number of correct")
plt.ylabel("Probability")
plt.tight_layout()
plt.legend()
plt.show()
```

### **5.**
A biologist observes a rare bird species, which appears with probability $0.05$ per hour of observation.
What is the probability that the first sighting happens after 10 hours?

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

k = 10
p = 0.05

x_vals = np.arange(100)
answer = stats.geom.sf(k, p)
pmf = stats.geom.pmf(x_vals, p)
colors = ['C1' if x > k else 'C0' for x in x_vals]
plt.bar(x_vals, pmf, color=colors)
plt.axvline(k+0.5, color="C1", linestyle="--", label=f"P(X > {k}) = {answer:.4}")
plt.title(f"Geometric(p = {p})")
plt.xlabel("Bird sightings per hour")
plt.ylabel("Probability")
plt.tight_layout()
plt.legend()
plt.show()
```
