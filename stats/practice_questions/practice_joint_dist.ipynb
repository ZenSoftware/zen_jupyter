Suppose $P(A) = 0.4$, $P(B) = 0.5$, and $P(A \cap B) = 0.2$.

Find:
- a) $P(A \cup B)$  
- b) $P(A^c \cap B)$  
- c) $P(A \mid B)$  
- d) Are A and B independent?

**Answer**
- a) $P(A)+P(B)-P(A \cap B) = 0.4 + 0.5 - 0.2 = 0.7$
- b) $0.6 \times 0.5 = 0.3$
- c) $\frac{0.3}{0.5} = 1.5$
- d) Yes because $P(A) \cdot P(B) = P(A \cap B)$
