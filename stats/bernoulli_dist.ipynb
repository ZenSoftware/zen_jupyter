# Bernoulli distribution
The Bernoulli distribution is the simplest possible distribution:  It models a single experiment that can result in only two possible outcomes - typically labeled:
- 1 = success
- 0 = failure

### Probability mass function:
- $P(X = 1) = p$
- $P(X = 0) = 1 âˆ’ p$

$p \in [0, 1]$

### Mean and Variance
- Mean: $E[X] = p$
- Variance: $V(X) = p(1-p)$

### Intuition and Examples
A Bernoulli trial is any experiment with a yes/no outcome:
- Flipping a coin (heads = 1, tails = 0)
- Asking: "Will the next dice roll be a 6?" (success = 1, failure = 0)
- Checking if a machine part is defective
