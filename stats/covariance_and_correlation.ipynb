## ðŸ“Œ Covariance: Definition & Meaning  
### **Formal definition**
$$
\operatorname{Cov}(X,Y)=\mathbb{E}\big[(X-\mu_X)(Y-\mu_Y)\big]
$$

- Measures **joint dependence** between $X$ and $Y$.  
- If $X = Y$, this reduces to variance:
  $$
  \operatorname{Var}(X)=\operatorname{Cov}(X,X)
  $$

### **Intuition**
- Look at how deviations from the mean line up:
  - If large $X-\mu_X$ tends to accompany large $Y-\mu_Y$, covariance is **positive**.  
  - If large $X-\mu_X$ tends to accompany *negative* $Y-\mu_Y$, covariance is **negative**.  
  - If thereâ€™s no consistent relationship, covariance is **near zero**.

### **Geometric picture**
- Imagine a cloud of points in the $xy$-plane:
  - **Tight, slanted ellipse** â†’ large covariance  
  - **Wide, less slanted ellipse** â†’ moderate covariance  
  - **Circular/symmetric cloud** â†’ covariance â‰ˆ 0  
- Slope + tightness = strength of covariance.

---

## ðŸ§® Useful Algebraic Formula  
Expanding the definition yields:
$$
\operatorname{Cov}(X,Y)=\mathbb{E}[XY]-\mathbb{E}[X]\mathbb{E}[Y]
$$

This is often easier to compute.

### **Independence**
- If $X$ and $Y$ are independent:
  $$
  \mathbb{E}[XY]=\mathbb{E}[X]\mathbb{E}[Y]
  $$
  so covariance = 0.

### **But the reverse is NOT true**
- Zero covariance does **not** imply independence.

---

## ðŸ§ª Example: Dependent but Zero Covariance  
The video gives a clean counterexample:

- $X$ uniform on $\{-1,0,1\}$  
- $Y = X^2$

These variables are **dependent**, but:
$$
\operatorname{Cov}(X,Y)=0
$$

Why?  
The positive and negative contributions cancel due to symmetry (even function over symmetric domain).

## ðŸ“ Variance of a Sum  
A key identity:
$$
\operatorname{Var}(X+Y)=\operatorname{Var}(X)+\operatorname{Var}(Y)+2\operatorname{Cov}(X,Y)
$$

- If $X$ and $Y$ are independent â†’ covariance = 0 â†’ variances add.  
- If dependent â†’ extra covariance term appears.

## ðŸ”— Correlation: Normalized Covariance  
### **Definition**
$$
\operatorname{Corr}(X,Y)=\frac{\operatorname{Cov}(X,Y)}{\sigma_X \sigma_Y}
$$

### **Why normalize?**
- Covariance changes if you scale units (e.g., feet â†’ inches).  
- Correlation removes scale effects, giving a **dimensionless** measure.

### **Properties**
- Always between $-1$ and $1$.  
- Invariant under linear transformations:
  $$
  \operatorname{Corr}(aX+b,\; cY+d)=\operatorname{Corr}(X,Y)
  $$

# âœ… What correlation measures that covariance does not

### **1. Correlation measures *strength and direction* of a linear relationship on a standardized scale.**

Covariance tells you whether two variables move together, but its magnitude is meaningless on its own because it depends on the units and scales of the variables.

Correlation fixes this by **standardizing** covariance:

$$
\rho_{XY} = \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y}
$$

This standardization gives correlation two special properties:

- It is **unitless**  
- It always lies between **âˆ’1 and 1**

So correlation answers a question covariance cannot:

> **How strong is the linear relationship, relative to the variability of each variable?**
