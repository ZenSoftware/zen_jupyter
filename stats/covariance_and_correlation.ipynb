## üéØ Big Picture  
- Covariance and correlation quantify **how two random variables vary together**.  
- They are central to probability, statistics, machine learning, PCA, regression, and high‚Äëdimensional data analysis.  
- Correlation is essentially a **normalized covariance**.
- Covariance can classify 3 types of relationships
  - Relationships with positive trends
  - Relationships with negative trends
  - No relationship because there is no trend

---

## üìå Covariance: Definition & Meaning  
### **Formal definition**
$$
\operatorname{Cov}(X,Y)=\mathbb{E}\big[(X-\mu_X)(Y-\mu_Y)\big]
$$

- Measures **joint dependence** between $X$ and $Y$.  
- If $X = Y$, this reduces to variance:
  $$
  \operatorname{Var}(X)=\operatorname{Cov}(X,X)
  $$

### **Intuition**
- Look at how deviations from the mean line up:
  - If large $X-\mu_X$ tends to accompany large $Y-\mu_Y$, covariance is **positive**.  
  - If large $X-\mu_X$ tends to accompany *negative* $Y-\mu_Y$, covariance is **negative**.  
  - If there‚Äôs no consistent relationship, covariance is **near zero**.

### **Geometric picture**
- Imagine a cloud of points in the $xy$-plane:
  - **Tight, slanted ellipse** ‚Üí large covariance  
  - **Wide, less slanted ellipse** ‚Üí moderate covariance  
  - **Circular/symmetric cloud** ‚Üí covariance ‚âà 0  
- Slope + tightness = strength of covariance.

---

## üßÆ Useful Algebraic Formula  
Expanding the definition yields:
$$
\operatorname{Cov}(X,Y)=\mathbb{E}[XY]-\mathbb{E}[X]\mathbb{E}[Y]
$$

This is often easier to compute.

### **Independence**
- If $X$ and $Y$ are independent:
  $$
  \mathbb{E}[XY]=\mathbb{E}[X]\mathbb{E}[Y]
  $$
  so covariance = 0.

### **But the reverse is NOT true**
- Zero covariance does **not** imply independence.

---

## üß™ Example: Dependent but Zero Covariance  
The video gives a clean counterexample:

- $X$ uniform on $\{-1,0,1\}$  
- $Y = X^2$

These variables are **dependent**, but:
$$
\operatorname{Cov}(X,Y)=0
$$

Why?  
The positive and negative contributions cancel due to symmetry (even function over symmetric domain).

---

## üìê Variance of a Sum  
A key identity:
$$
\operatorname{Var}(X+Y)=\operatorname{Var}(X)+\operatorname{Var}(Y)+2\operatorname{Cov}(X,Y)
$$

- If $X$ and $Y$ are independent ‚Üí covariance = 0 ‚Üí variances add.  
- If dependent ‚Üí extra covariance term appears.

---

## üîó Correlation: Normalized Covariance  
### **Definition**
$$
\operatorname{Corr}(X,Y)=\frac{\operatorname{Cov}(X,Y)}{\sigma_X \sigma_Y}
$$

### **Why normalize?**
- Covariance changes if you scale units (e.g., feet ‚Üí inches).  
- Correlation removes scale effects, giving a **dimensionless** measure.

### **Properties**
- Always between $-1$ and $1$.  
- Invariant under linear transformations:
  $$
  \operatorname{Corr}(aX+b,\; cY+d)=\operatorname{Corr}(X,Y)
  $$

## üß† Why This Matters  
- Covariance and correlation form the backbone of:
  - PCA  
  - SVD  
  - Regression  
  - High‚Äëdimensional pattern discovery  
- They quantify structure in data and reveal relationships between variables.
