# Geometric Distribution
The Geometric distribution models the number of trials you need to perform in a sequence of independent Bernoulli trials (each with probability *p*) until the first success occurs.  This makes it one of the most elegant and intuitive probability distributions in statistics.

There are actually **two common versions**, and both appear in litterature.

### Two Equivalent but Distinct Definitions
**Number of trials untill the first success**
- Support: $ k = 1, 2, 3, ... $
- PMF: $ P(X=k) = (1-p)^{k-1} p $
- Interpretation: "The first success happens on trial k."

**Number of failures before the first success**
- Support: $ k = 0, 1, 2, ... $
- PMF: $ P(X=k) = (1-p)^k p $
- Interpretation: "There are k failures before the first success"

Both describe the same physical process; they just count differently

### Key Properties
| Property | "Trials until success" version | "Failures before success" version |
|---|---|---|
| Mean | $ \frac{1}{p} $ | $ \frac{1-p}{p} $ |
| Variance | $ \frac{1-p}{p^2} $ | $ \frac{1-p}{p^2} $ |
| Mode | 1 | 0 |
| Memorylessness | Yes | Yes |

### Memoryless
The Geometric distribution is **the only discrete distribution that is memoryless**.  A distribution is **memoryless** if knowing that the event hasn't happened yet, tells you nothing about how much longer you'll have to wait.  Formally:
$$
P(X>m+n \mid X>m) = P(X>n)
$$
This says:
- If you've already waited *m* trials with no success,
- the probability you must wait *another* n trials
- is exactly the same as if you were starting from scratch.

The geometric distribution has this property due to the fact that it models repeated **independent Bernoulli trials**, each with:
- success probability $p$
- failure proability $1-p$

The key word is *independent*.  Independence means that, what happened before **does not change** the probability of success on the next trial.  So if you've already failrs 5 times, the chance you fail again on the next trial is still $1-p$.  And the chance you fail for the next 5 trials is still $(1-p)^5$.  That's exactly the same probability you'd have if you were starting fresh.

### Intuition and Examples
A geometric distribution describes situations like:
- Tossing a coin until the first heads appears
- Rolling a die until you get a 3
- Repeating a quality-control test until the first defect is found
- Trying to connect to a server until the first successful handshake

The key assumptions:
- Trials are independent
- Each trial has two outcomes (success/failure)
- Succes probability *p* is constant across trials

The geometric distribution is the **simplest waiting-time model** in probability.  It's the discrete analogue of the exponential distribution and forms the foundation for:
- Queueing theory
- Reliability engineering
- Markov chains
- Negative binomial distributions (sum of geometric variables)

### Plot

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

p = 0.3
xs = np.arange(1, 20)
pmf = stats.geom.pmf(xs, p)

fig, ax = plt.subplots()
ax.bar(xs, pmf)
ax.set_title(f"Geometric(p={p})")
ax.set_xlabel("Number of trials until the first success")
ax.set_ylabel("Proability")
ax.set_xticks(xs)
plt.show()
```

### Cumulative Distribution Function
For a geometric distribution with success probability $p$, the cumulative distribution function (CDF) gives the probability that the number of trials until the first success is **at most** $k$.

The CDF measures the chance that you succeed on or before trial $k$.
Since the only way this *doesn't* happen is if you fail the first $k$ trials, the CDF is:
$$
F(k) = P(X \leq k) = 1-(1-p)^k
$$

The term $(1-p)^k$ is the probability of **k consecutive failures**, so subtracting that from 1 will give you the probability that the number of trials until the first success is **at most** $k$.
