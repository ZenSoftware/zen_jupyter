# Geometric Distribution
The Geometric distribution models the number of trials you need to perform in a sequence of independent Bernoulli trials (each with probability *p*) until the first success occurs.  This makes it one of the most elegant and intuitive probability distributions in statistics.

There are actually **two common versions**, and both appear in litterature.

### Two Equivalent but Distinct Definitions
**Number of trials untill the first success**
- Support: $ k = 1, 2, 3, ... $
- PMF: $ P(X=k) = (1-p)^{k-1} p $
- Interpretation: "The first success happens on trial k."

**Number of failures before the first success**
- Support: $ k = 0, 1, 2, ... $
- PMF: $ P(X=k) = (1-p)^k p $
- Interpretation: "There are k failures before the first success"

Both describe the same physical process; they just count differently

### Key Properties
| Property | "Trials until success" version | "Failures before success" version |
|---|---|---|
| Mean | $ \frac{1}{p} $ | $ \frac{1-p}{p} $ |
| Variance | $ \frac{1-p}{p^2} $ | $ \frac{1-p}{p^2} $ |
| Mode | 1 | 0 |
| Memorylessness | Yes | Yes |

The **memoryless property** is especially famous:
$$
P(X>m+n | X>m) = P(X>n)
$$
This means the process "forgets" how long you've already waited.

### Intuition and Examples
A geometric distribution describes situations like:
- Tossing a coin until the first heads appears
- Rolling a die until you get a 3
- Repeating a quality-control test until the first defect is found
- Trying to connect to a server until the first successful handshake

The key assumptions:
- Trials are independent
- Each trial has two outcomes (success/failure)
- Succes probability *p* is constant across trials

The geometric distribution is the **simplest waiting-time model** in probability.  It's the discrete analogue of the exponential distribution and forms the foundation for:
- Queueing theory
- Reliability engineering
- Markov chains
- Negative binomial distributions (sum of geometric variables)

### Plot

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

p = 0.2
x = np.arange(1, 20)
pmf = stats.geom.pmf(x, p=p)

fig, ax = plt.subplots()
ax.bar(x, pmf)
ax.set_title(f"Geometric(p={p})")
ax.set_xlabel("Number of trials until the first success")
ax.set_ylabel("Proability density")
ax.set_xticks(x)
plt.show()
```
