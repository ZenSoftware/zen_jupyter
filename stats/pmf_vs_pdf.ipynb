# PMF vs PDF

For discrete variables (PMF): the height *is* the probability.

For continuos variables (PDF): the height is *not* the probability - only area is.

### Why PMFs behave differently
A **PMF** assigns probability to **individual points**:
$$
P(X=k) = p(k)
$$
And these values must satisfy:
- $ 0 \leq p(k) \leq 1 $
- $ \sum_k p(k) = 1 $

So the "height" of the PMF at a point **is literally the probability** of that point.

Example: a fair die
$$
p(1) = p(2) = ... = p(6) = \frac{1}{6}
$$
That height **is** the probability.

### Why the PDFs behave differently
A **PDF** assigns density, not probability:
$$
P(X=x) = 0
$$
Even if:
$$
f(x) = 2, 10, 1000
$$
The PDF can be tall because it's not a probability - it's a **rate of probabilty accumulation**.

Probability comes from area:
$$
P(a \leq X \leq b) = \int_a^b f(x) dx
$$

- PDF = height
- Probability = area
- A PDF can be > 1
- A CDF must be between 0 and 1
- Probability at a single point is always 0
- Only areas matter

A PDF tells you **how concentrated the probability is around a point**, not the probability *at* that point.
