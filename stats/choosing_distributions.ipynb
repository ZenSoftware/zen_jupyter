# Choosing Between Binomial, Poisson and Normal Distributions
| Distribution | When to Use It | Key Conditions |
|---|---|---|
| Binomial | Counting successes out of a fixed number of trials | Fixed n, two outcomes (success/failure), constant p, independent trials |
| Poisson | Counting rare events over time/space with no fixed number of trials | Events occur independently, at a constant average rate $\lambda$, events are rare |
| Normal | Modeling countinous, symmetric, bell-shaped data OR approximating Binomial/Poisson | Continous variable, roughly symmetric; OR large *n* for approximation |

### When to Use Each Distribution
#### Binomial Distribution
Use it when:
- You have a **fixed number of trials** *n*
- Each trial has **two outcomes** (success/failure)
- Probability of success *p* is **constant**
- Trials are **independent**

Examples:
- Number of defective items in a batch of 100
- Number of students who pass an exam out of 30
- Number of heads in 10 coin flips

**Why:** Binomial describes "binary data from a finite sample"

### Poisson Distribution
Use it when:
- You are counting **events in a continous interval** (time, area, volume)
- Events occur **independently**
- Events occur at a **constant average rate** $\lambda$
- Events are **rare** relative to the size of the interval

Examples:
- Number of emails arriving in an hour
- Number of accidents at an intersection per month
- Number of decay events in radioactive material

**Why:** Poisson describes "binary data from an infinite sample" and counts events in a population or interval

**Relationship to Binomial:** Poisson is a **limit of the Binomial** when:
- *n* is large
- *p* is small
- $\lambda=np$ is moderate

### Normal Distribution
Use it when:
- The variable is **continuous**
- The distribution is **symmetric and bell-shaped**
- You are modeling natural variation (height, test scores, measurement error)

Examples:
- IQ scores
- Measurement noise
- Human heights

**Why:** Normal describes continuous data with a symmetric bell shape

**Normal as an Approximation**

Normal can approximate:
- **Binomial** when $np > 5$ and $n(1-p) > 5$
- **Poisson** when $\lambda$ is large (typically $\lambda>10$)

This is why the Normal distribution is so widely used.

### How to Choose in Practice
Here's a simple flow:

Step 1: Is the variable continuous?
- Yes $\rightarrow$ Normal
- No $\rightarrow$ Go to Step 2

Step 2: Are you counting successes out of a fixed number of trials?
- Yes $\rightarrow$ Binomial
- No $\rightarrow$ Go to Step 3

Step 3: Are you counting events in time/space with no fixed number of trials?
- Yes $\rightarrow$ Poisson
- No $\rightarrow$ Re-axamine assumptions

### A Few Subtle Points
**Binomial vs Poisson**
- Use **Binomial** when you know the number of trials.
- Use **Poisson** when the number of opportunities is effectively unlimited or unknown.

**Poisson vs Normal**
- Use **Poisson** for small counts (0,1,2,3...)
- Use **Normal** when counts are large and symmetric

**Normal vs Binomial**
- Use **Normal** when the Binomial is too large to compute directly and conditions for approximation hold.
