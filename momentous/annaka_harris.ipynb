

**Annaka Harris is an American writer and editor whose work focuses on neuroscience, meditation, and the philosophy of mind.** She is best known for exploring the mysteries of consciousness and for making complex scientific ideas accessible to general audiences.   [bing.com](https://www.bing.com/search?q=annaka%20harris&filters=sid%3A%22f12bf38e-1b74-4764-2c5f-f54f15045c6b%22)

---

### 🧠 Who she is  
- **Writer & thinker** specializing in **consciousness**, **free will**, **panpsychism**, and the **hard problem of mind**.  
- **Author** of the *New York Times* bestseller **_Conscious: A Brief Guide to the Fundamental Mystery of the Mind_** (2019).   [Wikipedia](https://en.wikipedia.org/wiki/Annaka_Harris)  
- **Author of the children’s book _I Wonder_** (2013), which introduces themes of curiosity and uncertainty.   [Wikipedia](https://en.wikipedia.org/wiki/Annaka_Harris)  
- **Editor and consultant** for science writers, especially in neuroscience and physics.   [Annaka Harris](https://annakaharris.com/)  
- **Co‑founder of Project Reason**, a nonprofit promoting scientific education (2007).   [Wikipedia](https://en.wikipedia.org/wiki/Annaka_Harris)  

---

### 📚 Notable works  
- **_Conscious: A Brief Guide to the Fundamental Mystery of the Mind_** — explores the nature of consciousness, free will, and the possibility of panpsychism.   [Wikipedia](https://en.wikipedia.org/wiki/Annaka_Harris)  
- **_I Wonder_** — a children’s book about curiosity and the nature of reality.   [Wikipedia](https://en.wikipedia.org/wiki/Annaka_Harris)  
- **Editor of Sam Harris’s long‑form essay/book _Lying_** (2011).   [Wikipedia](https://en.wikipedia.org/wiki/Annaka_Harris)  
- **_LIGHTS ON_ (2025)** — an audio documentary series on neuroscience and consciousness.   [Wikipedia](https://en.wikipedia.org/wiki/Annaka_Harris)

### 🌟 Why she’s influential  
Annaka Harris stands out for her ability to bridge **science**, **philosophy**, and **mindfulness**, offering clear, elegant explanations of some of the most difficult questions in cognitive science. Her work appeals to readers interested in consciousness, meditation, and the nature of subjective experience.
